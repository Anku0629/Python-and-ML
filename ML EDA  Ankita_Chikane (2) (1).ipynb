```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv("data2 (1).csv")
df
```

```python
# step 1 
# find and replace missing values with NaN

df["Python"].replace("missing",np.nan,inplace=True)

df["Machine Learning"].replace("?",np.nan,inplace=True)

df["Age"].replace("?",np.nan,inplace=True)
df["Age"].replace("missing",np.nan,inplace=True)

df["AI"].replace("missing",np.nan,inplace=True)

df["NLP"].replace("?",np.nan,inplace=True)

df["Cv"].replace("?",np.nan,inplace=True)
df["Cv"].replace("missing",np.nan,inplace=True)
```

```python
df
```

```python
df.info()
```

```python
# step 2
# change datatype

df["Python"]=df["Python"].astype("float")
df["Machine Learning"]=df["Machine Learning"].astype("float")
df["Age"]=df["Age"].astype("float")
df["AI"]=df["AI"].astype("float")
df["NLP"]=df["NLP"].astype("float")
df["Cv"]=df["Cv"].astype("float")
```

```python
pmean=df["Python"].mean()
mlmean=df["Machine Learning"].mean()
amean=df["Age"].mean()
aimean=df["AI"].mean()
nlpmean=df["NLP"].mean()
cmean=df["Cv"].mean()
```

```python
# step 4
# fill null values

df["Python"].fillna(pmean,inplace=True)
df["Machine Learning"].fillna(mlmean,inplace=True)
df["Age"].fillna(amean,inplace=True)
df["AI"].fillna(aimean,inplace=True)
df["NLP"].fillna(nlpmean,inplace=True)
df["Cv"].fillna(cmean,inplace=True)
```

```python
df
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv("cars.csv")
df
```

```python
df.info()
```

```python
df.head()
```

```python
# Features and targets should be separated
# Features and targets should be numeric in nature
# Features and targets should be in proper shape and dimensions
```

```python
feature=df.iloc[:,:-1]
target=df.iloc[:,-1]
```

```python
feature
```

```python
target
```

```python
feature.replace("?",np.nan,inplace=True)
target.replace("?",np.nan,inplace=True)
```

```python
from sklearn.impute import SimpleImputer
si1=SimpleImputer(missing_values=np.nan,strategy="mean")
feature[["normalized-losses","horsepower"]]=si1.fit_transform(feature[["normalized-losses","horsepower"]])
```

```python
feature
```

# Outliers

# Skewness

```python
# If skewness is 0 no need to change
# If skewness is negative dont change
# If skewness highly positive then reduce it by applying log and square root method
# Those skewness is high but if it is highly correlated with target data then no need to change
```

```python
#For selecting specific datatype from dataframe(feature) select_dtype method is used as follows----

col=feature.select_dtypes(["int64","float64"]).columns
```

```python
# To know amount of skewness
```

```python
from scipy.stats import skew
```

```python
for i in feature[col]:
    print(i)
    print(skew(feature[i]))
    
    plt.figure
    sns.distplot(feature[i])
    plt.show()
```
