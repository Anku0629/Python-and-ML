```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score,classification_report
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.tree import DecisionTreeClassifier


import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv(r"C:\Users\admin\bank.csv")
```

```python
x=df.iloc[:,:-1]
y=df["deposit"]
```

```python
x
```

```python
y
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.30,random_state=50)
```

```python
#Making objects of each class

lreg=LogisticRegression()
knn=KNeighborsClassifier()
svm=SVC()
dt=DecisionTreeClassifier()
```

```python
#Function of MYMODEL

def mymodel(model):
    model.fit(xtrain,ytrain)
    ypred=model.predict(xtest)
    ac=accuracy_score(ytest,ypred)
    cr=classification_report(ytest,ypred)
    print(f"Accuracy: {ac} \n Classification report: {cr}")
```

```python
mymodel(lreg)
```

```python
mymodel(knn)
```

```python
mymodel(svm)
```

```python
mymodel(dt)
```

```python
# Checking overfitting of data
```

```python
dt.score(xtrain,ytrain)
```

### HyperTunning DT

```python
#Hyper parameter with impurity checker Gini

for i in range(1,31):
    dt1=DecisionTreeClassifier(max_depth=i)
    dt1.fit(xtrain,ytrain)
    ypred=dt1.predict(xtest)
    ac=accuracy_score(ytest,ypred)
    print(f"max Depth:{i}  accuarcy:{ac}")
```

# Max Deth Hyperparameter

```python
#Building model with final value of max depth as 8

dt2=DecisionTreeClassifier(max_depth=8)
mymodel(dt2)
```

```python
# Lets check overfitting scenario
dt2.score(xtrain,ytrain)
```

# min_samples_split

```python
for i in range(2,50):
    dt4=DecisionTreeClassifier(min_samples_split=i)
    dt4.fit(xtrain,ytrain)
    ypred=dt4.predict(xtest)
    ac=accuracy_score(ytest,ypred)
    print(f"Min Sample Split:{i}  accuarcy:{ac}")
```

```python
dt4=DecisionTreeClassifier(min_samples_split=43)
mymodel(dt4)
```

```python
dt4.score(xtrain,ytrain)
```

# min_samples_leaf

```python
for i in range(1,51):
    dt5=DecisionTreeClassifier(min_samples_leaf=i)
    dt5.fit(xtrain,ytrain)
    ypred=dt5.predict(xtest)
    ac=accuracy_score(ytest,ypred)
    print(f"Min Sample leaf:{i}  accuarcy:{ac}")
```

```python
dt5=DecisionTreeClassifier(min_samples_leaf=42)
mymodel(dt5)
```

```python
dt5.score(xtrain,ytrain)
```
