```python
#Que = show grade column ....
# avg>=80............A
# avg<80 and >70..............B
# avg<70 and >60........B+
# otherwise.........C
#avg=total/2
```

```python
def avg(marks):
    avg=marks
    if avg>=80:
        return "A"
    elif avg<80 and avg>70:
        return "B"
    elif avg<70 and avg>60:
        return "B+"
    else:
        return "C"
    
```

```python
df["Grade"]=df["Total"].apply(avg)
```
